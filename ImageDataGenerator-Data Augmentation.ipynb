```
cell_types = os.listdir('/content/drive/MyDrive/UITM/SEM 4/FYP/herlev_pap_smear/abnormal_light_dysplastic')
print('class=',len(cell_types))
```

```
cell_df=pd.DataFrame(data=cell_types, columns=['cell'])
print(cell_df.head())
```

```
path='/content/drive/MyDrive/UITM/SEM 4/FYP/herlev_pap_smear/abnormal_light_dysplastic'

im_size = 224

images=[]
for f in cell_types:
        img = cv2.imread(path + '/' + f)  # reading that image as array
        #print(img)  # will get the image as an array
        img = cv2.resize(img, (im_size, im_size))
        images.append(img)
```
